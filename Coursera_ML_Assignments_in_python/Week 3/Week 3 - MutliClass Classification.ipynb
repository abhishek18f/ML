```python
def sigmoid(z):                #sigmoid function
    return 1/(1+np.exp(-z))
```

```python
def h(theta, X):             #hypothesis function
    return sigmoid(np.dot(X, theta))
```

```python
def cost_func(theta, X, y):
    h1 = h(theta, X)
    J = float(-(np.dot(y.T, np.log(h1)) + np.dot((1-y).T, np.log(1 - h1)))/m) + lambd*((theta*theta).sum() - (theta[0]*theta[0]))/(2*m)
    return J[0]

def grad(theta , X ,y ):
    h1 = h(theta, X)
    grad = np.dot(X.T , (h1 - y))/m + lambd*theta/m
    grad[0] = np.dot(X.T , (h1 - y))[0]/m
    return grad
```

```python
lambd = 0.1
initial_theta  = np.zeros((X.shape[1], 1))
alpha = 0.001
theta = initial_theta

def optim(X,y,theta):
    for i in range(100000):
        theta -= alpha*grad(X,y,theta)
        if(i%10000 ==0):
            print("Iteration " + str(i) + "  Cost: " + str(cost_func(X,y,theta)) )
    return h(theta, X)    
```

```python
def y_k(temp_y, i):
    index = temp_y==i
    #print(index)
    temp_y[:] =0
    temp_y[index] = 1
    return temp_y
```

```python
#optim(X,y,theta).shape
hypo =[]
for i in range(10):
    print("------ " +str(i)+ "-------" )
    temp_y = np.copy(y)
    temp_y = y_k(temp_y,i)
    temp = opt.fmin_tnc(func = cost_func, 
                        x0 = theta,fprime = grad, 
                        args = (X, temp_y))
    hypo.append(temp[0])
```

```python
hypothesis = np.zeros((m, 10))
for i in range(10):
    print("------ " +str(i)+ "-------" )
    temp_y = np.copy(y)
    temp_y = y_k(temp_y,i)
    hypothesis[:, i:i+1] = optim(X,temp_y, theta)
    temp_y
```
