```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn import datasets
from scipy.io import loadmat
```

```python
data2 = loadmat('ex6data2.mat')
X2, y2 = data2['X'] , data2['y'].ravel()
m = y2.size
clf2 = svm.SVC(C=100)
clf2.fit(X2,y2)
```
